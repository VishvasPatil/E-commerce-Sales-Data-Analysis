## Extracting Month from Order Date

```python
# Extract month directly without creating a new column
df['Month'] = df['Order Date'].dt.month
```

```python
df.dtypes
```

```python
df["Quantity Ordered"]=df["Quantity Ordered"].astype(int)
df["Price Each"]=df["Price Each"].astype(float)
```

## Product wise Sales

```python
df.groupby(["Product"])["Quantity Ordered"].sum().plot(kind="bar")
```

```python
count_df=df.groupby(["Product"]).agg({"Quantity Ordered":"sum","Price Each":"mean"})
```

```python
count_df=count_df.reset_index()
```

```python
Products=count_df["Product"].values
```

```python
import warnings

# Filter out the specific warning
warnings.filterwarnings("ignore", message="FixedFormatter should only be used together with FixedLocator")
```

```python
fig,ax1=plt.subplots()
ax2=ax1.twinx()

ax1.bar(count_df["Product"],count_df['Quantity Ordered'])
ax2.plot(count_df["Product"],count_df['Price Each'],color="red")
ax1.set_xticklabels(Products,rotation="vertical")

ax1.set_ylabel("Order Count")
ax2.set_ylabel("Average Price")
```

```python
df_duplicats=df[df['Order ID'].duplicated(keep=False)]
```

```python
df_duplicats.groupby(['Order ID'])["Product"]
```

```python
df_duplicats.groupby(['Order ID'])["Product"].apply(lambda x:','.join(x))
```

```python
Group_Products=df_duplicats.groupby(['Order ID'])["Product"].apply(lambda x:','.join(x)).reset_index()
```

```python
Group_Products.rename(columns={"Product":"group_products"})
```

```python
Group_Products=Group_Products.rename(columns=({"Product":"group_products"}))
```

```python
df_duplicates=df_duplicats.merge(Group_Products,how='left',on='Order ID')
```

```python
df_No_duplicates=df_duplicates.drop_duplicates(subset=["Order ID"])
```
